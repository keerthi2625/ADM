```python
from sklearn.preprocessing import StandardScaler
```

```python
df3='/content/titanic.csv'
```

```python
data = pd.read_csv(df3)
```

```python
data['Age']=data['Age'].fillna(data['Age'].mean())
```

```python
standard_scaler = StandardScaler()# Initialize the StandardScaler
```

```python
data['Age_standard'] = standard_scaler.fit_transform(data[['Age']])# Standardize the Age column
```

```python
print(data[['Age','Age_standard']].head())
```

```python
data['Age']=data['Age'].fillna(data['Age'].median())# Handle missing values in the Age column by imputing with the median
```

```python
print(data[['Age','Age_standard']].head())
```

```python
data['Age']=data['Age'].fillna(data['Age'].mode()[0])# Handle missing values in the Age column by imputing with the mode
```

```python
print(data[['Age','Age_standard']].head())
```
